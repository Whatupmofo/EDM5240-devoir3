```python
url = "https://opencanada.blob.core.windows.net/opengovprod/resources/1d15a62f-5656-49ad-8c88-f40ce689d831/grants.csv?sr=b&sp=r&sig=ZuPi4SHube/v4kzq/yZTj/V07GLn2od33oRfug19nGc%3D&sv=2015-07-08&se=2019-03-17T14%3A07%3A19Z"
```

```python
req = requests.get(url).content
```

```python
divulgation=pan.read_csv(StringIO(req.decode("utf-8")), low_memory=False)
```

<h3 style="background-color:lightgreen;">Cette ligne a pour fonction de faire la liste de tous les programmes du ministère. On y constate que c'est le Fonds du Canada pour les périodiques qui a octroyé le plus de subventions. Je commence à manquer d'imagination pour les gifs</h3>

```python
periodique = divulgation.nom_programme.str.contains("Fonds du Canada pour les périodiques", na=False)
divulgation[periodique]
```

<h3 style="background-color:lightgreen;">Et voici le sous-ensemble pour le Fonds du Canada pour les périodiques</h3>

```python
divulgation.valeur_accord = pan.to_numeric(divulgation.valeur_accord.str.replace("$","").str.replace(",",""))
```

<h3 style="background-color:lightgreen;">Ici, c'est pour faire un petit tri dans la collone du montant pour transformer le type de la valeur en int plutôt qu'en str.</h3>

```python
divulgation[periodique].groupby("nom_legal_beneficiaire").valeur_accord.sum().sort_values(ascending=False)
```

<h3 style="background-color:lightgreen;">On peut dire sans se tromper que le groupe TVA s'en dire plutôt bien.</h3>

```python
divulgation[periodique].groupby("province_beneficiaire").valeur_accord.sum().sort_values(ascending=False) / divulgation[periodique].valeur_accord.sum() * 100
```
